```python
import numpy as np
from sklearn.cluster import MeanShift
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import style
plt.rcParams["figure.figsize"] = "11, 8"                  #IMPORTATN LINE TO ENLARGE GRAPHICAL PLOT
style.use("ggplot")
```

```python
#defining some random centres as the three points
centers = [[1,1,1],[5,5,5],[3,10,10]]
```

```python
#defining the number of randomly generated data points to be 100
X, _ = make_blobs(n_samples = 1000, centers = centers, cluster_std = 1.5) 
```

```python
ms = MeanShift()                     #using sciki-learns mean-shift and then training
ms.fit(X)
```

```python
labels = ms.labels_
cluster_centers = ms.cluster_centers_

print(cluster_centers)
n_clusters_ = len(np.unique(labels))
print("Number of estimated clusters:", n_clusters_)
```

```python
#defining the colours and the plotting of the figure
colors = 10*['r','g','b','c','k','y','m']
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
for i in range(len(X)):
    ax.scatter(X[i][0], X[i][1], X[i][2], c=colors[labels[i]], marker='o')

ax.scatter(cluster_centers[:,0],cluster_centers[:,1],cluster_centers[:,2],
            marker="x",color='k', s=150, linewidths = 5, zorder=10)

plt.show()
```
